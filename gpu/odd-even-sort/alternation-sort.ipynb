### <a name=introduction>Introduction</a>

A circuit model of computation has compute elements connected by wires, typically in layers, such that computation occurs through the flow of data through the computational elemments connected by the wires.

A question arises as to how to sort in a network. The $n$ input wires have values on them, and those values should appear, after having traversed the network, on $n$ output wires with values ascending or descending according to the wire ennumeration. For instance, 

<pre>
      +-------+
 2 ---|       |--- 7
      |       |
 4 ---|       |--- 4
      |  SORT |
 1 ---|       |--- 2
      |       |
 7 ---|       |--- 1
      +-------+
</pre>

The only computational element allowed is the swap. Given any two wires, for simplicity in the same layer, put the values in order by allowing them to pass through or by swapping them across the wires.

This is a short note to jot down a proof for a particular network that has $n$ layers, using $n\,(n-1)/2$ swap elements, with the elements only swaping neighboring wires, where $n$ is even. I call it _Alternation Sort_, but I should probably research what its real name is.

Why this is interesting is that,

- First off, it shows a speed up of a factor of $\log n$ because of parallel sorting. This is not nothing. We can do depth $\log^2 n$ sorts, which puts this to shame. Still it's not nothing.
- Second, the easiest sorts of depth $O(n)$ can be modeled after serial insertion sort, or selection sort. These can them be simply modified to get a depth of $2n$. This sort cuts that in half.
- Third, the depth is optimal for a network where the swaps are only with neighbors. Consider the input of all zeros and a one at the bottom wire. To bring it up to the top wire, where it must end, will require $n-1$ swaps.


### <a name=alternationsort>Alternation Sort</a>

A simple method to sort is given $n$ even is the have $n$ layers which only swap among neighboring wires. The layers alternate of two types. The even layers create $n/2$ pairs of neighboring wires starting with the top wire. The odd layers create $n/2-1$ pairs of neighboring wires, not including the top and bottom wires.

<pre>
     ---+-----+------
        |     |
     ---+--+--+--+---
           |     |
     ---+--+--+--+---
        |     |
     ---+-----+------
</pre>

### <a name=proof>Proof</a>

__Proof__: The proof that this works is by induction, two wires are a time. Suppose that for $n-2$ wires this circuilt works, with $n-2$ layers.

Using the 0-1 theorem, consider some application of 0's and 1's at the input. Consider the two 1's, if they exist, closest to the top. Consider all possible configurations for these two 1's, and how they move upwards. After the first two layers they have moved upwards so that the one's and zero's below both will never interact with these two ones. Therefore this reduces to $n-2$ elements that will sort in the remaining $n-2$ layers. (Note that it does reduce to $n-2$ wires, as the top two wires will have the 1's that we are following.)
